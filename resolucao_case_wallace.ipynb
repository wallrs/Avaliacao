```python
# Importandos as bibliotecas necessária.
import pandas as pd 
import numpy as np
import warnings 

# Configurando para visualizar o máximo de colunas e ignorar erros. 
pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

```python
# Importando a base de perfil_eleitorado como perfil_df.
perfil_df = pd.read_csv('perfil_eleitorado_2020.csv', encoding= "latin-1", sep = ';')
```

```python
# Importando a base de SP_turno_1 como resultado_df.
resultado_df = pd.read_csv('SP_turno_1.csv', encoding= "latin-1", sep = ';')
```

```python
# Visualizando a base de dado perfil_eleitorado e consultando quantas linhas e colunas contém nessa base de dados. 
display(perfil_df)
display(perfil_df.shape)
```

```python
# Visualizando a base de dado SP_turno_1 e consultando quantas linhas e colunas contém nessa base de dados.
display(resultado_df)
display(resultado_df.shape)
```

```python
# TRATAMENTO DO DATASET PERFIL_ELEITORAL_2020.csv

# Filtrando o dataset perfil_eleitorado pela coluna SG_UF para buscamos somente os perfil dos eleitores do estado de SP.
perfil_df = perfil_df.loc[(perfil_df['SG_UF'] == "SP")]
perfil_df.info()

```

```python
# Visualisando as 5 primeiras linhas para tratamento dessa base. 
perfil_df.head()
```

```python
# Irémos excluir as colunas HH_GERACAO, ANO_ELEICAO, e  as coluna CD_MUN_SIT_BIOMETRIA e DS_MUN_SIT_BIOMETRIA também serão deletada,
# nesse ano o estado suspendeu o cadastramento para eleição de 2020 por causa da pandemia. Essas colunas não vamos precissar para análise.
# fonte: https://www.camarasjc.sp.gov.br/noticias/7087/justica-dispensa-cadastro-biometrico-para-as-eleicoes-de-2020
perfil_df = perfil_df.drop(columns=['CD_GRAU_ESCOLARIDADE','CD_ESTADO_CIVIL','CD_GENERO','CD_MUNICIPIO',
                                    'HH_GERACAO','ANO_ELEICAO','CD_MUN_SIT_BIOMETRIA','DS_MUN_SIT_BIOMETRIA'])

# Consultando se as colunas foram deletadas.
perfil_df.info()
```

```python
# Consultamos que há algumas colunas que estão com tipo de data númerico, porem são colunas de idenficação, vamos converter para tipo texto (string/object).
perfil_df[['NR_ZONA','CD_FAIXA_ETARIA']] = perfil_df[['NR_ZONA','CD_FAIXA_ETARIA']].astype(str)

# Validando se as colunas foram convertida.
perfil_df.info()
```

```python
## Vamos verificar se há valores ausente ou nulo.
perfil_df.isna().sum()
```

```python
# TRATAMENTO DA DATASET SP_TURNO_1

# Visualisando as 5 primeiras linhas para tratamento dessa base. 
resultado_df.head()
```

```python
# Nesse banco de dados vamos deletar algumas colunas que serão desnecessária para nossa análise
resultado_df = resultado_df.drop(columns=['DT_GERACAO','DS_AGREGADAS','HH_GERACAO','ANO_ELEICAO','CD_TIPO_ELEICAO','NM_TIPO_ELEICAO','CD_PLEITO',
                                          'DT_PLEITO','NR_TURNO','CD_ELEICAO','DS_ELEICAO','DT_BU_RECEBIDO','CD_TIPO_URNA','CD_MUNICIPIO','CD_CARGO_PERGUNTA',
                                          'DS_TIPO_URNA','CD_CARGA_1_URNA_EFETIVADA','CD_CARGA_2_URNA_EFETIVADA','CD_FLASHCARD_URNA_EFETIVADA','DS_CARGO_PERGUNTA_SECAO',
                                          'DT_CARGA_URNA_EFETIVADA','QT_ELEITORES_BIOMETRIA_NH','DT_EMISSAO_BU','NR_JUNTA_APURADORA','NR_TURMA_APURADORA'])                          

# Consultando se as colunas foram deletadas.
resultado_df.info()
```

```python
# E também temos algumas colunas com tipo de dado incorreto.
# Vamos coverter as colunas numerica (int) para texto (str).
resultado_df['NR_ZONA'] = resultado_df['NR_ZONA'].astype(str)
resultado_df['NR_VOTAVEL'] = resultado_df['NR_VOTAVEL'].astype(str)
resultado_df['NR_SECAO'] = resultado_df['NR_SECAO'].astype(str)
resultado_df['NR_LOCAL_VOTACAO'] = resultado_df['NR_LOCAL_VOTACAO'].astype(str)
resultado_df['NR_PARTIDO'] = resultado_df['NR_PARTIDO'].astype(str)
resultado_df['CD_TIPO_VOTAVEL'] = resultado_df['CD_TIPO_VOTAVEL'].astype(str)
resultado_df['NR_URNA_EFETIVADA'] = resultado_df['NR_URNA_EFETIVADA'].astype(str)

# Validando se os tipos de dados foram convertidos
resultado_df.info()
```

```python
# Visualizando o dataset
resultado_df.head()

# As colunas NR_PARTIDO, SG_PARTIDO e NM_PARTIDO contém valores nulos (-1 e #NULO#),
# iremos substituir os valores nulos por Branco e Nulos.
# As colunas com esses valores nulos se trata das siglas e nome do partidos.
resultado_df.loc[resultado_df.NM_VOTAVEL == 'Branco',['NR_PARTIDO','SG_PARTIDO','NM_PARTIDO']] = 'Branco'
resultado_df.loc[resultado_df.NM_VOTAVEL == 'Nulo',['NR_PARTIDO','SG_PARTIDO','NM_PARTIDO']] = 'Nulo'

```

```python
#Verificando se valores foram substituidos 
display(resultado_df[['DS_TIPO_VOTAVEL','NR_PARTIDO','SG_PARTIDO','NM_PARTIDO']])
```

```python
# Consultando se há vazio
display(resultado_df.isna().sum())
```

```python
# Para as colunas DT_ABERTURA e DT_ENCERRAMENTO vamos substituir os valores vázios por 15/11/1020 07:00:00 e 15/11/2020 17:00:00
# Por se tratar do horário padrão de aberturar e encerramento da votação
resultado_df['DT_ABERTURA'].fillna('15/11/2020 07:00:00', inplace=True)
resultado_df['DT_ENCERRAMENTO'].fillna('15/11/2020 17:00:00', inplace=True)

# Temos 13 valores vázio na coluna NM_VOTAVEL, iremos deletar essas linhas númores baixo.
resultado_df = resultado_df.dropna()

# consultando se ainda há valores vázios para colunas DT_ABERTURA e DT_ENCERRAMENTO
display(resultado_df.isna().sum())
```

```python
# E as Colunas DT_ABERTURA E DT_ENCARRAMENTO tem data e horas dentro do mesmo campo, 
# iremos criar mais duas colunas para inserimos os horarios separadamente para futuras análise.
resultado_df['HH_ABERTURA'] = resultado_df['DT_ABERTURA'].str.slice(start=11, stop=19)
resultado_df['DT_ABERTURA'] = resultado_df['DT_ABERTURA'].str.slice(start=0, stop=10)
resultado_df['HH_ENCERRAMENTO'] = resultado_df['DT_ENCERRAMENTO'].str.slice(start=11, stop=19)
resultado_df['DT_ENCERRAMENTO'] = resultado_df['DT_ENCERRAMENTO'].str.slice(start=0, stop=10)

# Visualizando se as duas colunas foram criada.
resultado_df.head()
```

```python
# Consultando qual canditado foram mais votado por município.
mais_votados = resultado_df.groupby('NM_MUNICIPIO')[['SG_PARTIDO','NM_VOTAVEL','QT_VOTOS']].max()
display(mais_votados)
```

```python
# Vamos consultar quantidade do eleitorado pelo sexo, ensino, e idade e salvar em variaveis.
contagem_genero = perfil_df['DS_GENERO'].value_counts()
contagem_idade = perfil_df['DS_FAIXA_ETARIA'].value_counts()
contagem_ensino = perfil_df['DS_GRAU_ESCOLARIDADE'].value_counts()
contagem_pcd = perfil_df['QT_ELEITORES_DEFICIENCIA'].value_counts()
```

```python
# Visualizando os resultado.
print('\nContagem de eleitores por sexo:')
print(contagem_genero)

print('\nContagem de eleitores pela idade:')
print(contagem_idade)

print('\nContagem de eleitores pelo grau de escolaridade:')
print(contagem_ensino)
```

```python
# Vamos salvar as bases tratada.
perfil_df.to_csv('Resul_perfil_df_.csv', index=False, encoding='utf-8')
resultado_df.to_csv('Resul_resultado_df_.csv',index=False, encoding='utf-8')
```
